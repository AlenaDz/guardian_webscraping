```python
import pandas as pd
from requests_html import HTMLSession
```

```python
# Creating an empty dataframe with columns

guardian_subjects = pd.DataFrame(columns = ["Rank", "Institution", "Guardian score/100", "Satisfied with course", "Satisfied with teaching", "Satisfied with feedback", "Student to staff ratio", "Spend per student/10", "Average entry tariff", "Value added score/10", "Career after 6 months"])
guardian_subjects
```

```python
# Defining the web scraping function
def guardian(html, subject):
    
# Scraping raw HTML
    session = HTMLSession()
    page = session.get(html)
        
# Extracting table lines    
    results = []
    
    for link in page.html.find('td'):
      results.append(link.text)

# Deleting unneeded information 
    key = "Other universities that taught this subject"
    
    if key in results:
        stop = results.index(key)
        results_clean = results[:stop]
    else: 
        results_clean = results

# Cleaning results from empty strings (also important because some empty strings represent ads on the page)
    final = list(filter(lambda a: a != '', results_clean))
    
# Creating a list for each row that we are interested in
    rank2018 = []
    institution = []
    guardian100 = []
    satisfied_with_course = []
    satisfied_with_teaching = []
    satisfied_with_feedback = []
    student_to_staff = []
    spend_per_student = []
    avg_entry_tariff = []
    value_added_score = []
    career_6m = []

    index = 0
    while index <= len(final)-13:
      rank2018.append(final[index])
      index +=13

    index = 1
    while index <= len(final)-12:
      institution.append(final[index])
      index +=13

    index = 2
    while index <= len(final)-11:
      guardian100.append(final[index])
      index +=13

    index = 3
    while index <= len(final)-10:
      satisfied_with_course.append(final[index])
      index +=13

    index = 4
    while index <= len(final)-9:
      satisfied_with_teaching.append(final[index])
      index +=13   

    index = 5
    while index <= len(final)-8:
      satisfied_with_feedback.append(final[index])
      index +=13

    index = 6
    while index <= len(final)-7:
      student_to_staff.append(final[index])
      index +=13

    index = 7
    while index <= len(final)-6:
      spend_per_student.append(final[index])
      index +=13

    index = 8
    while index <= len(final)-5:
      avg_entry_tariff.append(final[index])
      index +=13

    index = 9
    while index <= len(final)-4:
      value_added_score.append(final[index])
      index +=13

    index = 10
    while index <= len(final)-3:
      career_6m.append(final[index])
      index +=13

# Combining all lists into a dataframe, adding the Subject and Year
    data_2018 = pd.DataFrame([rank2018, institution, guardian100, satisfied_with_course, satisfied_with_teaching, satisfied_with_feedback, student_to_staff, spend_per_student, avg_entry_tariff, value_added_score, career_6m])
    data_2018 = data_2018.transpose()
    data_2018.columns = ["Rank", "Institution", "Guardian score/100", "Satisfied with course", "Satisfied with teaching", "Satisfied with feedback", "Student to staff ratio", "Spend per student/10", "Average entry tariff", "Value added score/10", "Career after 6 months"]
    data_2018["Year"] = 2018
    data_2018["Subject"] = subject

# Appending the data for the current subject to all subjects
    guardian_new = guardian_subjects.append(data_2018).reset_index(drop = True)
    return guardian_new
```

```python
# Preparing the list of subjects which are at the same time part of URL

subjects = ['accounting-finance', 'american-studies', 'anatomy-physiology', 'anthropology', 'architecture', 'art', 'biosciences', 
            'building-and-town-and-country-planning', 'business-management-marketing', 'chemistry', 'classics-ancient-history',
            'computer-science-information', 'criminology', 'dentistry', 'design-crafts', 'drama-dance', 'earth-marine-sciences',
            'economics', 'education', 'chemical-engineering', 'civil-engineering','electronic-electrical-engineering',
            'general-engineering', 'materials-mineral-engineering', 'mechanical-engineering', 'english-creative-writing',
            'fashion-textiles', 'film-production-photography', 'forensic-science-archaeology', 'geography-environmental-studies',
            'health-professions', 'history', 'history-of-art', 'hospitality-event-management-tourism',
            'journalism-publishing-public-relations', 'law', 'mathematics', 'media-film-studies', 'medicine',
            'modern-languages-linguistics', 'music', 'nursing-midwifery', 'pharmacy-pharmacology', 'philosophy', 'physics', 'politics',
            'psychology', 'religious-studies-and-theology', 'social-policy-administration', 'social-work', 'sociology', 'sports-science',
            'veterinary-science']
```

```python
# Running the scraping funcation for all subjects

for x in subjects:
    html = f'https://www.theguardian.com/education/ng-interactive/2017/may/16/university-guide-2018-league-table-for-{x}'
    guardian_new = guardian(html, x)
    guardian_subjects = guardian_new

guardian_subjects
```

```python
# Reordering columns

final_guardian = guardian_subjects[["Year", "Subject", "Rank", "Institution", "Guardian score/100", "Satisfied with course", "Satisfied with teaching", "Satisfied with feedback", "Student to staff ratio", "Spend per student/10", "Average entry tariff", "Value added score/10", "Career after 6 months"]]
```
